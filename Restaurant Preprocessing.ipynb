```python
#Adding a date to each input and formatting it for manipulation
date_1 = "2019-01-02 12:00:00"
     
date_lst = []
count = -1

for each in df['Unnamed: 0']:
    count += 1
    if count % 16 == 0:
        date_1 = datetime.datetime.strptime(str(date_1),"%Y-%m-%d %H:%M:%S")
        date_1 = date_1 + datetime.timedelta(days=1)
        date_lst.append(date_1)
    else:
        date_1 = datetime.datetime.strptime(str(date_1),"%Y-%m-%d %H:%M:%S")
        date_lst.append(date_1)
       
    
dates= []
for each in date_lst:
    each = each.strftime('%Y-%m-%d')
    dates.append(each)
    
df['Date']=dates
df.to_csv(r'C:\Users\harr1\Desktop\Weather_7.csv')
```

```python
#Convert each column to numeric

weather_df['Date_num']= pd.to_datetime(weather_df.Date_num,infer_datetime_format=True)
weather_df['apparentTemperature'] = pd.to_numeric(weather_df['apparentTemperature'], errors='coerce')
weather_df['cloudCover'] = pd.to_numeric(weather_df['cloudCover'], errors='coerce')
weather_df['dewPoint'] = pd.to_numeric(weather_df['dewPoint'], errors='coerce')
weather_df['humidity'] = pd.to_numeric(weather_df['humidity'], errors='coerce')
weather_df['ozone'] = pd.to_numeric(weather_df['ozone'], errors='coerce')
weather_df['pressure'] = pd.to_numeric(weather_df['pressure'], errors='coerce')
weather_df['temperature'] = pd.to_numeric(weather_df['temperature'], errors='coerce')
weather_df['time'] = pd.to_numeric(weather_df['time'], errors='coerce')
weather_df['uvIndex'] = pd.to_numeric(weather_df['uvIndex'], errors='coerce')
weather_df['visibility'] = pd.to_numeric(weather_df['visibility'], errors='coerce')
weather_df['windBearing'] = pd.to_numeric(weather_df['windBearing'], errors='coerce')
weather_df['windGust'] = pd.to_numeric(weather_df['windGust'], errors='coerce')
weather_df['windSpeed'] = pd.to_numeric(weather_df['windSpeed'], errors='coerce')
weather_df.info()

#Merging weather and restaurant datasets
everything_raw= pd.merge(all_data, weather_df, left_on='Date', right_on='Date_num')
everything_raw['Menu Group'].unique()
everything = everything_raw.drop(['Date', 'Date_num'], axis=1)
everything.info()
```

```python
#changing time from timestamp to days since open

everything_raw.time.min()
everything_raw['time_adjust'] = [abs((1546516800 - x)/86400) for x in everything_raw.time]
everything_raw.time_adjust
everything = everything_raw.drop('time', axis=1)
everything
```

```python
#Preprocessing to turn all categorical data into numerical data
from sklearn.preprocessing import LabelEncoder
everything['Day_encoded'] = LabelEncoder().fit_transform(everything['DOTW'])
everything['weather'] = LabelEncoder().fit_transform(everything['icon'])
everything['summary_num'] = LabelEncoder().fit_transform(everything['summary'])
everything = everything.drop(['summary','DOTW', 'icon', 'Date', 'Date_num'], axis = 1)

everything.info()
```

```python
#Adding a feature for the hours the restaurant is open per day of the week

hours =[]
for each in everything['Day_encoded']:
    if each == 0:
        hours.append(6)
    elif each == 1:
        hours.append(5)
    elif each == 2:
        hours.append(12)
    elif each ==3:
        hours.append(10.5)
    elif each ==4:
        hours.append(6)
    elif each == 5:
        hours.append(5)
    else:
        hours.append(5)

everything['hours']=hours

#Adding a feature for the month the data is from
month = []
for each in everything['time_adjust']:
    if each <= 28:
        month.append(1)
    elif 28<each<=56:
        month.append(2)
    elif 56<each<=87:
        month.append(3)
    elif 88<each<=117:
        month.append(4)
    elif 118<each<148:
        month.append(5)
    elif 149<each<178:
        month.append(6)
    else:
        month.append(7)
        
everything['month']=month
```
